```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.spatial.distance import cdist
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================
# Phase 1 - Data Preprocessing & Feature Engineering
# ==============================

# Load dataset (Upload to Colab before running)
from google.colab import files
uploaded = files.upload()  # Upload your CSV here
df = pd.read_csv(next(iter(uploaded.keys())))
print("=== First 5 Rows of Data ===")
print(df.head())
```

```python
print("=== Handling Missing Numeric Values ===")
imputer = SimpleImputer(strategy='mean')
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])
# Handle missing categorical values
print("=== Handling Missing Categorical Values ===")
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
print("=== Feature Engineering (Haversine Distance) ===")
def haversine(lat1, lon1, lat2, lon2):
    R = 6371
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    return R * (2 * np.arcsin(np.sqrt(a)))

if all(col in df.columns for col in ['Restaurant_lat', 'Restaurant_long', 'Customer_lat', 'Customer_long']):
    df['Distance_km'] = haversine(df['Restaurant_lat'], df['Restaurant_long'],
                                   df['Customer_lat'], df['Customer_long'])
```

```python
print("=== Encoding Categorical Features ===")
encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_df = pd.DataFrame(encoder.fit_transform(df[cat_cols]),
                          columns=encoder.get_feature_names_out(cat_cols))
df = pd.concat([df.drop(cat_cols, axis=1), encoded_df], axis=1)

# Normalize all features
print("=== Scaling Features ===")
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
# --- K-Means Clustering ---
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

# Re-apply preprocessing and scaling to ensure df_scaled is available
# Assuming df is already loaded from a previous cell
# Handle missing numeric values
imputer = SimpleImputer(strategy='mean')
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])
# Handle missing categorical values
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Feature Engineering (Haversine Distance) - assuming relevant columns exist
def haversine(lat1, lon1, lat2, lon2):
    R = 6371
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    return R * (2 * np.arcsin(np.sqrt(a)))

if all(col in df.columns for col in ['Restaurant_latitude', 'Restaurant_longitude', 'Delivery_location_latitude', 'Delivery_location_longitude']):
    df['Distance_km'] = haversine(df['Restaurant_latitude'], df['Restaurant_longitude'],
                                   df['Delivery_location_latitude'], df['Delivery_location_longitude'])


# Encoding Categorical Features
encoder = OneHotEncoder(sparse_output=False, drop='first')
# Select only object type columns for encoding
cat_cols_for_encoding = df.select_dtypes(include=['object']).columns
encoded_df = pd.DataFrame(encoder.fit_transform(df[cat_cols_for_encoding]),
                          columns=encoder.get_feature_names_out(cat_cols_for_encoding))
df = pd.concat([df.drop(cat_cols_for_encoding, axis=1), encoded_df], axis=1)


# Normalize all features
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)


print("=== Running Elbow Method ===")
distortions = []
K_range = range(2, 6)  # Reduced for speed
for k in K_range:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(df_scaled)
    distortions.append(sum(np.min(cdist(df_scaled, kmeans.cluster_centers_, 'euclidean'), axis=1)) / df_scaled.shape[0])

plt.figure(figsize=(6, 4))
plt.plot(K_range, distortions, marker='o')
plt.title('Elbow Method For Optimal K')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()

optimal_k = 3
print(f"Using Optimal K = {optimal_k}")

kmeans = KMeans(n_clusters=optimal_k, n_init=10, random_state=42)
df['KMeans_Cluster'] = kmeans.fit_predict(df_scaled)

plt.figure(figsize=(6, 4))
sns.scatterplot(x=df_scaled.iloc[:, 0], y=df_scaled.iloc[:, 1], hue=df['KMeans_Cluster'], palette='viridis')
plt.title("K-Means Clusters")
plt.show()
```

```python
# --- Hierarchical Clustering ---
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import numpy as np

print("=== Running Hierarchical Clustering ===")
sample_data = df_scaled.iloc[:200]  # Only first 200 rows for dendrogram
linked = linkage(sample_data, method='ward')

plt.figure(figsize=(8, 5))
dendrogram(linked, truncate_mode='lastp', p=12, show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram (First 200 Samples)')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import tensorflow as tf
from tensorflow import keras
```

```python
from google.colab import files

uploaded = files.upload()
df = pd.read_csv(next(iter(uploaded.keys())))
```

```python
# Remove leading/trailing spaces from column names
df.columns = df.columns.str.strip()

# Convert 'Time_taken(min)' to numeric, coercing errors
df['Time_taken(min)'] = pd.to_numeric(df['Time_taken(min)'], errors='coerce')

# Drop rows where 'Time_taken(min)' is NaN after coercion
df.dropna(subset=['Time_taken(min)'], inplace=True)

# Create a binary target variable based on the median time
median_time = df['Time_taken(min)'].median()
df['Time_taken_binary'] = (df['Time_taken(min)'] > median_time).astype(int)


y = df['Time_taken_binary']
X = df.drop(columns=['Time_taken(min)', 'Time_taken_binary']) # Drop both original and new target columns

# Encode categorical columns
cat_cols = X.select_dtypes(include=['object']).columns
encoder = OneHotEncoder(sparse_output=False, drop='first')
X_encoded = pd.DataFrame(encoder.fit_transform(X[cat_cols]),
                         columns=encoder.get_feature_names_out(cat_cols))
X = pd.concat([X.drop(columns=cat_cols), X_encoded], axis=1)

# Scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
model = keras.Sequential([
    keras.layers.Dense(32, activation='relu', input_shape=(X_train.shape[1],)),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=20, batch_size=16, validation_split=0.2, verbose=1)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred = model.predict(X_test)

# If model outputs probabilities (e.g., neural network), get class labels
if y_pred.ndim > 1 and y_pred.shape[1] > 1:
    y_pred = y_pred.argmax(axis=1)  # pick class with highest probability

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, average='macro'))
print("Recall:", recall_score(y_test, y_pred, average='macro'))
print("F1-score:", f1_score(y_test, y_pred, average='macro'))
```
